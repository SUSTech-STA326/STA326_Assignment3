```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# 定义根目录和文件名
root_directory = './results'
filename = 'hit_ratio.csv'

# 准备数据
data = {}

# 遍历根目录下的所有文件夹
for folder in os.listdir(root_directory):
    folder_path = os.path.join(root_directory, folder)
    file_path = os.path.join(folder_path, filename)
    
    # 检查文件是否存在
    if os.path.isfile(file_path):
        # 读取CSV文件
        df = pd.read_csv(file_path)
        
        # 检查文件中是否有数据
        if not df.empty:
            data[folder] = df

# 创建绘图
plt.figure(figsize=(10, 6))

# 遍历每个文件的数据并绘制
for folder, df in data.items():
    epochs = range(1, len(df) + 1)  # 创建一个 epoch 的范围
    plt.plot(epochs,df, label=folder)

# 设置图表标题和标签
plt.title('MovieLens')
plt.xlabel('Epoch')
plt.ylabel('HR@10')
plt.legend()
plt.grid(True)

# 显示图表
plt.show()
```
