```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
HR_all = [
    f"../result/ml-1m_{model}_HR10{'_3Layer' if model == 'MLP' else ''}.csv" for model in ["GMF", "MLP", "Neu_MF"]
]
HRs = pd.DataFrame()
for file in HR_all:
    df = pd.read_csv(file, index_col=0)
    HRs = pd.concat([HRs, df], axis=1)
```

```python
plt.figure(figsize=(10, 6), dpi = 160)
for column in HRs.columns:
    plt.plot(HRs.index, HRs[column], label=column)

plt.xlabel('Epoch')
plt.ylabel('HR@10')
plt.title('HR@10 over epochs')
plt.legend()
plt.grid(True)
plt.savefig('../result/HR10_all.png')
plt.show()
```

```python
NDCG_all = [
    f"../result/ml-1m_{model}_NDCG10{'_3Layer' if model == 'MLP' else ''}.csv" for model in ["GMF", "MLP", "Neu_MF"]
]
NDCGs = pd.DataFrame()
for file in NDCG_all:
    df = pd.read_csv(file, index_col=0)
    NDCGs = pd.concat([NDCGs, df], axis=1)
```

```python
plt.figure(figsize=(10, 6), dpi = 160)
for column in NDCGs.columns:
    plt.plot(NDCGs.index, NDCGs[column], label=column)

plt.xlabel('Epoch')
plt.ylabel('NDCG@10')
plt.title('NDCG@10 over epochs')
plt.legend()
plt.grid(True)
plt.savefig('../result/NDCG10_all.png')
plt.show()
```

```python
HR_MLPs = pd.DataFrame()
for layer in range(5):
    df = pd.read_csv(f"../result/ml-1m_MLP_HR10_{layer}Layer.csv", index_col=0) \
        .rename(columns={"ml-1m_MLP": f"ml-1m_MLP_{layer}Layer"})
    HR_MLPs = pd.concat([HR_MLPs, df], axis=1)
```

```python
plt.figure(figsize=(10, 6), dpi = 160)
for column in HR_MLPs.columns:
    plt.plot(HR_MLPs.index, HR_MLPs[column], label=column)

plt.xlabel('Epoch')
plt.ylabel('HR@10')
plt.title('HR@10 over epochs')
plt.legend()
plt.grid(True)
plt.savefig('../result/HR10_MLPs.png')
plt.show()
```
