```python
import matplotlib.pyplot as plt
import numpy as np

# 数据示例
factors =  ['8','16','32','64']
hr_values = {
    'GMF': GMF_hr,
    'MLP': MLP_hr,
    'NeuMF': NCF_hr
}

            
ndcg_values = {
    'GMF': GMF_ndcg,
    'MLP': MLP_ndcg,
    'NeuMF': NCF_ndcg
}

# 创建图表
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

# 绘制 HR@10
for name, values in hr_values.items():
    ax[0].plot(factors, values, marker='o', label=name)
ax[0].set_title('HR@10')
ax[0].set_xlabel('Factors')
ax[0].set_ylabel('HR@10')
ax[0].grid(True)
ax[0].legend()

# 绘制 NDCG@10
for name, values in ndcg_values.items():
    ax[1].plot(factors, values, marker='o', label=name)
ax[1].set_title('NDCG@10')
ax[1].set_xlabel('Factors')
ax[1].set_ylabel('NDCG@10')
ax[1].grid(True)
ax[1].legend()

plt.tight_layout()
plt.show()
```
