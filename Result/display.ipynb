```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置Seaborn样式
sns.set(style="whitegrid")
```

```python
gmf = pd.read_csv("./GMF_metrics.csv")
mlp = pd.read_csv("./mlp_metrics.csv")
neumf = pd.read_csv("./GMF_metrics.csv")
```

```python
# 提取Hit Ratio和NDCG
gmf_hit = gmf['Hit Ratio']
gmf_ndcg = gmf['NDCG']
mlp_hit = mlp['Hit Ratio']
mlp_ndcg = mlp['NDCG']
neumf_hit = neumf['Hit Ratio']
neumf_ndcg = neumf['NDCG']

# 设置图表大小
plt.figure(figsize=(14, 6))

# 绘制Hit Ratio图表
plt.subplot(1, 2, 1)
plt.plot(gmf_hit, label='GMF', marker='o', linestyle='-', linewidth=2, markersize=6)
plt.plot(mlp_hit, label='MLP', marker='s', linestyle='--', linewidth=2, markersize=6)
plt.plot(neumf_hit, label='NeuMF', marker='^', linestyle='-.', linewidth=2, markersize=6)
plt.title('Hit Ratio@10', fontsize=16)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Hit Ratio', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 绘制NDCG图表
plt.subplot(1, 2, 2)
plt.plot(gmf_ndcg, label='GMF', marker='o', linestyle='-', linewidth=2, markersize=6)
plt.plot(mlp_ndcg, label='MLP', marker='s', linestyle='--', linewidth=2, markersize=6)
plt.plot(neumf_ndcg, label='NeuMF', marker='^', linestyle='-.', linewidth=2, markersize=6)
plt.title('NDCG@10', fontsize=16)
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('NDCG', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 调整布局并显示图表
plt.tight_layout()
plt.show()
```

```python
# 读取数据
ab0 = pd.read_csv("./mlp_ablation_0.csv")
ab1 = pd.read_csv("./mlp_ablation_1.csv")
ab2 = pd.read_csv("./mlp_ablation_2.csv")
ab3 = pd.read_csv("./mlp_ablation_3.csv")
ab4 = pd.read_csv("./mlp_ablation_4.csv")

# 计算均值
ablation_datasets = [ab0, ab1, ab2, ab3, ab4]
mean_values = []

for i, ab in enumerate(ablation_datasets):
    hit_ratio_mean = ab["Hit Ratio"].max()
    ndcg_mean = ab["NDCG"].max()
    mean_values.append((hit_ratio_mean, ndcg_mean))
    print(f"Dataset ab{i}: Hit Ratio Mean = {hit_ratio_mean}, NDCG Mean = {ndcg_mean}")
```
