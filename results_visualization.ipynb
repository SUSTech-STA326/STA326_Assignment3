```python

# 提取文件名中的评估数据
results = []
pattern = re.compile(r'^(gmf|mlp|neumf)_.*_Epoch(\d+)_HR([\d.]+)_NDCG([\d.]+)\.model$')

for name in lines:
    match = pattern.search(name)
    if match:
        model_type = match.group(1)
        epoch = int(match.group(2))
        hr = float(match.group(3))
        ndcg = float(match.group(4))
        results.append((model_type,epoch,hr,ndcg))
    else:
        print(name)
    
results.append(('gmf',18,0.6364,0.3651))
# 将结果转换为DataFrame
results_df = pd.DataFrame(results, columns=['Model Type','Epoch', 'Hit Ratio', 'NDCG'])


```

```python
# 排序数据以便正确的绘制
gmf = results_df[results_df['Model Type']=='gmf']
mlp = results_df[results_df['Model Type']=='mlp']
neumf = results_df[results_df['Model Type']=='neumf']

gmf = gmf.sort_values(by='Epoch')
mlp = mlp.sort_values(by='Epoch')
neumf = neumf.sort_values(by='Epoch')

# 可视化评估数据
plt.figure(figsize=(6, 4))
plt.plot(gmf['Epoch'], gmf['Hit Ratio'],  label='gmf',markersize=5)
plt.plot(mlp['Epoch'], mlp['Hit Ratio'], label='mlp',markersize=5)
plt.plot(neumf['Epoch'], neumf['Hit Ratio'], label='neumf',markersize=5)
plt.xlabel('Epoch')
plt.ylabel('HR@10')
plt.title('HR@10 vs. Epoch')
plt.legend()
plt.savefig("layer3_HR.png")
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(gmf['Epoch'], gmf['NDCG'], label='gmf', markersize=5)
plt.plot(mlp['Epoch'], mlp['NDCG'],label='mlp',markersize=5)
plt.plot(neumf['Epoch'], neumf['NDCG'],label='neumf',markersize=5)
plt.xlabel('Epoch')
plt.ylabel('NDCG@10')
plt.title('NDCG@10 vs. Epoch')
plt.legend()
# plt.savefig("layer3_NDCG.png")
plt.show()
```
