```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
gmf_hr = pd.read_csv('gmf_hr.csv')
mlp_hr = pd.read_csv('mlp_hr.csv')
neumf_hr= pd.read_csv('neumf_hr.csv')

gmf_ndcg = pd.read_csv('gmf_ndcg.csv')
mlp_ndcg = pd.read_csv('mlp_ndcg.csv')
neumf_ndcg=pd.read_csv('neumf_ndcg.csv')

neumf_h1=pd.read_csv("neumf_hr_1.csv")
neumf_h2=pd.read_csv("neumf_hr_2.csv")
neumf_h3=pd.read_csv("neumf_hr_3.csv")


neumf_n1=pd.read_csv("neumf_ndcg_1.csv")
neumf_n2=pd.read_csv("neumf_ndcg_2.csv")
neumf_n3=pd.read_csv("neumf_ndcg_3.csv")
```

```python
import pandas as pd

# 创建全为 0 的 Series
zero_row = pd.Series([0] * len(gmf_hr.columns), index=gmf_hr.columns)

# 在 DataFrame 的最前面添加全为 0 的行
gmf_hr = pd.concat([pd.DataFrame([zero_row]), gmf_hr], ignore_index=True)
mlp_hr = pd.concat([pd.DataFrame([zero_row]), mlp_hr], ignore_index=True)
neumf_hr = pd.concat([pd.DataFrame([zero_row]), neumf_hr], ignore_index=True)

gmf_ndcg = pd.concat([pd.DataFrame([zero_row]), gmf_ndcg], ignore_index=True)
mlp_ndcg = pd.concat([pd.DataFrame([zero_row]), mlp_ndcg], ignore_index=True)
neumf_ndcg = pd.concat([pd.DataFrame([zero_row]),neumf_ndcg], ignore_index=True)

neumf_h1 = pd.concat([pd.DataFrame([zero_row]), neumf_h1], ignore_index=True)
neumf_h2 = pd.concat([pd.DataFrame([zero_row]), neumf_h2], ignore_index=True)
neumf_h3 = pd.concat([pd.DataFrame([zero_row]),neumf_h3], ignore_index=True)

neumf_n1 = pd.concat([pd.DataFrame([zero_row]), neumf_n1], ignore_index=True)
neumf_n2 = pd.concat([pd.DataFrame([zero_row]), neumf_n2], ignore_index=True)
neumf_n3 = pd.concat([pd.DataFrame([zero_row]),neumf_n3], ignore_index=True)

```

```python
fig, ax = plt.subplots()
ax.plot(gmf_hr['Step'], gmf_hr['Value'], label='gmf')
ax.plot(mlp_hr['Step'], mlp_hr['Value'], label='mlp')
ax.plot(neumf_hr['Step'], neumf_hr['Value'], label='neumf')

ax.set(xlabel='Epoch', ylabel='Hit Ratio@10',
       title='performs between gmf, mlp and neumf')
ax.grid()
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(gmf_ndcg['Step'], gmf_ndcg['Value'], label='gmf')
ax.plot(mlp_ndcg['Step'], mlp_ndcg['Value'], label='mlp')
ax.plot(neumf_ndcg['Step'], neumf_ndcg['Value'], label='neumf')

ax.set(xlabel='Epoch', ylabel='NDCG@10',
       title='performs between gmf, mlp and neumf')
ax.grid()
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(neumf_n1['Step'], neumf_n1['Value'], label='layer1')
ax.plot(neumf_n2['Step'], neumf_n2['Value'], label='layer2')
ax.plot(neumf_n3['Step'], neumf_n3['Value'], label='layer3')
ax.plot(neumf_ndcg['Step'], neumf_ndcg['Value'], label='layer4')

ax.set(xlabel='Epoch', ylabel='NDCG@10',
       title='performs between different layers')
ax.grid()
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(neumf_h1['Step'], neumf_h1['Value'], label='layer1')
ax.plot(neumf_h2['Step'], neumf_h2['Value'], label='layer2')
ax.plot(neumf_h3['Step'], neumf_h3['Value'], label='layer3')
ax.plot(neumf_hr['Step'], neumf_hr['Value'], label='layer4')

ax.set(xlabel='Epoch', ylabel='Hit Ratio@10',
       title='performs between different layers')
ax.grid()
plt.legend()
plt.show()
```
