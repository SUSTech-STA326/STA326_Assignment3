```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
gmf_hr = pd.read_csv('gmf_hr.csv')
mlp_hr = pd.read_csv('mlp_hr.csv')
neumf_hr= pd.read_csv('neumf_hr.csv')

gmf_ndcg = pd.read_csv('gmf_ndcg.csv')
mlp_ndcg = pd.read_csv('mlp_ndcg.csv')
neumf_ndcg = pd.read_csv('neumf_ndcg.csv')
```

```python
import pandas as pd

# 创建全为 0 的 Series
zero_row = pd.Series([0] * len(gmf_hr.columns), index=gmf_hr.columns)

# 在 DataFrame 的最前面添加全为 0 的行
gmf_hr = pd.concat([pd.DataFrame([zero_row]), gmf_hr], ignore_index=True)
mlp_hr = pd.concat([pd.DataFrame([zero_row]), mlp_hr], ignore_index=True)
neumf_hr = pd.concat([pd.DataFrame([zero_row]), neumf_hr], ignore_index=True)

gmf_ndcg = pd.concat([pd.DataFrame([zero_row]), gmf_ndcg], ignore_index=True)
mlp_ndcg = pd.concat([pd.DataFrame([zero_row]), mlp_ndcg], ignore_index=True)
neumf_ndcg = pd.concat([pd.DataFrame([zero_row]),neumf_ndcg], ignore_index=True)
```

```python
fig, ax = plt.subplots()
ax.plot(gmf_hr['Step'], gmf_hr['Value'], label='gmf')
ax.plot(mlp_hr['Step'], mlp_hr['Value'], label='mlp')
ax.plot(neumf_hr['Step'], neumf_hr['Value'], label='neumf')

ax.set(xlabel='Epoch', ylabel='Hit Ratio@10',
       title='Hit Ratio: MLP performs slightly better than GMF after pretraining')
ax.grid()
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(gmf_ndcg['Step'], gmf_ndcg['Value'], label='gmf')
ax.plot(mlp_ndcg['Step'], mlp_ndcg['Value'], label='mlp')
ax.plot(neumf_ndcg['Step'], neumf_ndcg['Value'], label='neumf')

ax.set(xlabel='Epoch', ylabel='NDCG@10',
       title='NDCG: MLP performs slightly better than GMF after pretraining')
ax.grid()
plt.legend()
plt.show()
```
