```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 
gmf_df = pd.read_csv('GMF.csv')
mlp_df = pd.read_csv('MLP.csv')
neumf_df = pd.read_csv('NeuMF.csv')

gmf_df.insert(0, 'epoch', range(1, 31))
mlp_df['epoch'] = mlp_df.index % 30 + 1
neumf_df.insert(0, 'epoch', range(1, 31))

# 
combined_df = pd.concat([gmf_df, mlp_df, neumf_df])

# 
plt.figure(figsize=(12, 8))
sns.lineplot(x='epoch', y='HR@10', hue='Model', data=combined_df)
plt.title('Model Performance Comparison - HR@10')
plt.xlabel('Model')
plt.ylabel('HR@10')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('./HR.png',dpi=300)
plt.show()

plt.figure(figsize=(12, 8))
sns.lineplot(x='epoch', y='NDCG@10', hue='Model', data=combined_df)
plt.title('Model Performance Comparison - NDCG@10')
plt.xlabel('Model')
plt.ylabel('NDCG@10')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.tight_layout()
plt.savefig('./NDCG.png',dpi=300)
plt.show()
```
