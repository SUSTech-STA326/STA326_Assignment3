```python
import pickle

# 打开一个文件以二进制读取模式
with open('factor_8_logs.pkl', 'rb') as file:
    # 使用pickle.load()加载对象
    data8 = pickle.load(file)
with open('factor_16_logs.pkl', 'rb') as file:
    # 使用pickle.load()加载对象
    data16 = pickle.load(file)
with open('factor_32_logs.pkl', 'rb') as file:
    # 使用pickle.load()加载对象
    data32 = pickle.load(file)
with open('factor_64_logs.pkl', 'rb') as file:
    # 使用pickle.load()加载对象
    data64 = pickle.load(file)
print(data8)
```

```python
HRs = {}
NDCGs = {}
HRs['NCF'] = []
NDCGs['NCF'] = []
HRs['GMF'] = []
NDCGs['GMF'] = []
HRs['MLP'] = []
NDCGs['MLP'] = []

for i, (key, value) in enumerate(data8.items(), start=0):
    if key == 'logs':
        continue
    if key == 'HRs':
        HRs['NCF'].append(value['NCF'][-1])
        HRs['GMF'].append(value['GMF'][-1])
        HRs['MLP'].append(value['MLP'][-1])
    if key == 'NDCGs':
        NDCGs['GMF'].append(value['GMF'][-1])
        NDCGs['NCF'].append(value['NCF'][-1])
        NDCGs['MLP'].append(value['MLP'][-1])

for i, (key, value) in enumerate(data16.items(), start=0):
    if key == 'logs':
        continue
    if key == 'HRs':
        HRs['NCF'].append(value['NCF'][-1])
        HRs['GMF'].append(value['GMF'][-1])
        HRs['MLP'].append(value['MLP'][-1])
    if key == 'NDCGs':
        NDCGs['GMF'].append(value['GMF'][-1])
        NDCGs['NCF'].append(value['NCF'][-1])
        NDCGs['MLP'].append(value['MLP'][-1])

for i, (key, value) in enumerate(data32.items(), start=0):
    if key == 'logs':
        continue
    if key == 'HRs':
        HRs['NCF'].append(value['NCF'][-1])
        HRs['GMF'].append(value['GMF'][-1])
        HRs['MLP'].append(value['MLP'][-1])
    if key == 'NDCGs':
        NDCGs['GMF'].append(value['GMF'][-1])
        NDCGs['NCF'].append(value['NCF'][-1])
        NDCGs['MLP'].append(value['MLP'][-1])

for i, (key, value) in enumerate(data64.items(), start=0):
    if key == 'logs':
        continue
    if key == 'HRs':
        HRs['NCF'].append(value['NCF'][-1])
        HRs['GMF'].append(value['GMF'][-1])
        HRs['MLP'].append(value['MLP'][-1])
    if key == 'NDCGs':
        NDCGs['GMF'].append(value['GMF'][-1])
        NDCGs['NCF'].append(value['NCF'][-1])
        NDCGs['MLP'].append(value['MLP'][-1])

import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(15, 7))
plt.subplot(1, 2, 1)
for model in ['NCF', 'GMF', 'MLP']:
    plt.plot(['8', '16', '32', '64'], HRs[model], label=model)
plt.xlabel('factor')
plt.ylabel('HR')
plt.title('HR')
plt.legend()
plt.subplot(1, 2, 2)
for model in ['NCF', 'GMF', 'MLP']:
    plt.plot(['8', '16', '32', '64'], NDCGs[model], label=model)
plt.xlabel('factor')
plt.ylabel('NDCG')
plt.title('NDCG')
plt.legend()
plt.show()
```
