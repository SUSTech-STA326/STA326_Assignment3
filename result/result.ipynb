```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('ggplot')
```

# Experiment 1: Model Compare

```python
data = pd.read_csv("result.csv")
data['log loss'] = np.log(data['Train Loss'])
# 假设df是你的DataFrame
df = data[data['# Factor'] == 8]
models = df['Model']
iterations = df['Iteration']

# 创建一个图形和轴对象

fig, ax = plt.subplots(figsize=(8, 8))

# 假设有多个模型，我们可以通过groupby来分组并为每个模型绘制一条线
for model, group in df.groupby('Model'):
    ax.plot(group['Iteration'], group['log loss'], label=model)

# 添加图例
ax.legend(title='Model')

# 添加标题和轴标签
ax.set_title('Training Loss Over Iterations')
ax.set_xlabel('Iteration')
ax.set_ylabel('Log Train Loss')
ax.grid()
# 显示图表
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
# 假设有多个模型，我们可以通过groupby来分组并为每个模型绘制一条线
for model, group in df.groupby('Model'):
    ax.plot(group['Iteration'], group['HR@10'], label=model)

# 添加图例
ax.legend(title='Model')

# 添加标题和轴标签
ax.set_title('HR@10 Over Iterations')
ax.set_xlabel('Iteration')
ax.set_ylabel('HR@10')
ax.yaxis.grid(True)
# 显示图表
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
# 假设有多个模型，我们可以通过groupby来分组并为每个模型绘制一条线
for model, group in df.groupby('Model'):
    ax.plot(group['Iteration'], group['NDCG@10'], label=model)

# 添加图例
ax.legend(title='Model')

# 添加标题和轴标签
ax.set_title('NDCG@10 Over Iterations')
ax.set_xlabel('Iteration')
ax.set_ylabel('NDCG@10')
ax.grid()
# 显示图表
plt.show()
```
