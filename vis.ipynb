```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
result=pd.read_csv('result.csv',index_col=0)
result.head(2)
```

```python
# plt.rcParams['font.family'] = 'Arial'
plt.plot(np.arange(50),result[result['alias']=='gmf']['hit_ratio'],label='gmf',linewidth=2)
plt.plot(np.arange(50),result[result['alias']=='mlp']['hit_ratio'],label='mlp',linewidth=2)
plt.plot(np.arange(50),result[result['alias']=='neumf']['hit_ratio'],label='neumf',linewidth=2)
plt.xlabel('iteration',size=16)
plt.ylabel('HR@10',size=16)
plt.title('Movie Lens',size=16)
plt.legend(fontsize=16)
```

```python
plt.plot(np.arange(50),result[result['alias']=='gmf']['ndcg'],label='gmf',linewidth=2)
plt.plot(np.arange(50),result[result['alias']=='mlp']['ndcg'],label='mlp',linewidth=2)
plt.plot(np.arange(50),result[result['alias']=='neumf']['ndcg'],label='neumf',linewidth=2)
plt.xlabel('iteration',size=16)
plt.ylabel('ndcg@10',size=16)
plt.title('Movie Lens',size=16)
plt.legend(fontsize=16)
```
