```python
import os
import csv
import pandas as pd
import matplotlib.pyplot as plt

tasks = ["result/gmf.csv","result/mlp0.csv","result/mlp1.csv",
        "result/mlp2.csv","result/mlp3.csv","result/mlp4.csv",
        "result/neumf.csv"]

def read_csv_files(file_paths):
    data = {}
    for file_path in file_paths:
        df = pd.read_csv(file_path)
        key = os.path.basename(file_path).replace('.csv', '')
        data[key] = df
    return data

def plot_metrics(data):
    # First plot: gmf, mlp3, neumf metrics
    plt.figure()
    for key in ['gmf', 'mlp3', 'neumf']:
        if key in data:
            plt.plot(data[key]['Epoch'], data[key]['HR'], label=f'{key} HR@10')
            plt.plot(data[key]['Epoch'], data[key]['NDCG'], label=f'{key} NDCG@10')
    plt.xlabel('Epoch')
    plt.ylabel('Metrics')
    plt.title('Metrics for gmf, mlp3, neumf')
    plt.legend()
    plt.grid(True)
    plt.savefig('result/gmf_mlp3_neumf_metrics.png', dpi=400)
    plt.show()

    # Second plot: mlp0 to mlp4 HR@10
    plt.figure()
    for key in ['mlp0', 'mlp1', 'mlp2', 'mlp3', 'mlp4']:
        if key in data:
            plt.plot(data[key]['Epoch'], data[key]['HR'], label=f'{key}')
    plt.xlabel('Epoch')
    plt.ylabel('HR@10')
    plt.title('HR@10 for mlp0 to mlp4')
    plt.legend()
    plt.grid(True)
    plt.savefig('result/mlp0_to_mlp4_hr.png', dpi=400)
    plt.show()

    # Third plot: mlp0 to mlp4 NDCG@10
    plt.figure()
    for key in ['mlp0', 'mlp1', 'mlp2', 'mlp3', 'mlp4']:
        if key in data:
            plt.plot(data[key]['Epoch'], data[key]['NDCG'], label=f'{key}')
    plt.xlabel('Epoch')
    plt.ylabel('NDCG@10')
    plt.title('NDCG@10 for mlp0 to mlp4')
    plt.legend()
    plt.grid(True)
    plt.savefig('result/mlp0_to_mlp4_ndcg.png', dpi=400)
    plt.show()
```

```python
# Read the CSV files
data = read_csv_files(tasks)
# Plot the metrics
plot_metrics(data)
```
